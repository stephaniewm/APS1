```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#defindo os parametros
p = 0.434
g = 9.81
raio = 0.86
teta = math.radians(20)
t = np.arange(0,5, 0.001)

#equação diferencial
def Sistema(S, t):
    teta = S[0]
    w = S[1]
    dtetadt = w
    dwdt = (-g * math.sin(teta))/raio
    return [dtetadt, dwdt]

    
cond_inicial = [teta, 0]

#executando o odeint
solucao = odeint(Sistema, cond_inicial, t)

x = []
y = []
p = []

#criando a posiçao em relação aos eixos x e y
h=0
for teta in solucao[:,0]:
    x.append(raio * math.cos(teta))
    y.append(raio * math.sin(teta))

p.append(h)

#achando o periodo
periodo = []
for h in range(1, len(solucao[:,0])):
    if solucao[:,0][h-1]>solucao[:,0][h]:
        periodo.append(t[h])
        break
        
print(periodo)

#plotando os gráficos
plt.plot(t, solucao[:,0])
plt.title("Variação do ângulo por tempo")
plt.show()

plt.plot(y, x)
plt.axis([min(y), max(y), max(x), min(x)])
plt.title("Trajetória do pêndulo")
plt.axis("equal")
plt.show()


plt.plot(t, x)
plt.title("Posição no eixo x por tempo")
plt.show()

plt.plot(t, y)
plt.title("Posição no eixo y por tempo")
plt.show()


#usando a formula para calcular o periodo

periodoformula = 2 * math.pi * (raio/g)**0.5
print(periodoformula)
```
